```python
import numpy as np
import matplotlib.pyplot as plt
```

Se importan las librerías necesarias.

```python
L=1
k=1
c=1
N=50
```

Se añaden los parámetros.

```python
x=np.linspace(0, L, 100)
t=0.1
```

Dominio y tiempo.

```python
def f(x):
     return np.where(x<L/2, 2*k/L*x, 2*k/L*(L-x))
def An(n):
     n_pi=n*np.pi
     primer_termino=(4*k/(n_pi)**2)*np.sin(n_pi/2)
     return primer_termino
```

Se define f(x) y el coeficiente An.

```python
def u_xt(x,t,N_te=50):
     u=np.zeros_like(x)
     for n in range(1, N_te+1):
          an=An(n)
          u+=an*np.sin(n*np.pi*x/L)*np.cos(n*np.pi*c*t/L)
     return u
```

Se construye la solución u(x,t).
Se suman los primeros N números para aproximar la solución.

```python
# Evaluar y graficar la solución
u = u_xt(x, t, N)
plt.plot(x, u, label=f"t = {t}")
plt.title("Solución de la ecuación de onda")
plt.xlabel("x")
plt.ylabel("u(x,t)")
plt.grid(True)
plt.legend()
plt.show()
```
