```python
from sympy import *
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
x=sp.symbols('x')
lamb=sp.symbols('lambda')
y=sp.Function('y')
ode=sp.Eq(sp.Derivative(y(x),x,x)-lamb*y(x),0)
sol=sp.dsolve(ode)
#print(sol)

C1, C2 = sp.symbols('C1 C2')
X_sol = sol.rhs
#Condiciones
eq1 = sp.Eq(X_sol.subs(x, 0), 0)
eq2 = sp.Eq(X_sol.subs(x, 1), 0)

constants = sp.solve([eq1, eq2], (C1, C2))

n = sp.Symbol('n', integer=True, positive=True)
lambda_n = (n * sp.pi)**2
X_n = sp.sin(n * sp.pi * x)

print("\nValores propios y eigenfunciones:")
print(f"lambda_n = {lambda_n}")
print(f"X_n(x) = sin(n * pi * x)")
# Graficamos las primeras tres eigenfunciones
x_vals = np.linspace(0, 1, 100)
plt.figure(figsize=(10, 6))

for n_val in range(1, 4):
    X_func = sp.lambdify(x, sp.sin(n_val * sp.pi * x), 'numpy')
    plt.plot(x_vals, X_func(x_vals), label=f'n = {n_val},  = {(n_val*np.pi)**2:.2f}')

plt.title("Eigenfunciones para X'' - X = 0 con X(0) = X(1) = 0")
plt.xlabel("x")
plt.ylabel("X(x)")
plt.legend()
plt.grid(True)
plt.show()
```
